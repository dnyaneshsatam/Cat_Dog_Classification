### Let's randomly Test 

```python
import glob
import random
import os
```

```python
dataset_path = 'C:\\Users\\Dnyanesh\\Cat_Dog_Classification\\dataset\\test_set'

path = []
test_cls_label = {}
test_img_path = []
cls = ['cats','dogs']

for i,cd_class in enumerate(cls):
    path = glob.glob(os.path.join(dataset_path,cd_class,'*'))
    test_img_path = test_img_path + path
    test_cls_label.update({p:i for p in path})

#print(len(test_img_path))
#print(len(test_cls_label))   

random.shuffle(test_img_path)
random.shuffle(test_img_path)

## let's test results

correct_count = 0
actual_class = ''
prediction=''
total = 2000

for image_path in test_img_path:
    
    actual_class = test_cls_label[image_path]
    
    img = cv2.imread(image_path)
    img = cv2.resize(img,(100,100))
    img = img/255
    img = np.expand_dims(img,axis=0)
    prediction = model.predict_classes(img)
    
    
    if prediction == actual_class:
        correct_count = correct_count + 1
        
    actual_class = ''
    prediction = ''
    

print('Total Images Tested : ',total)
print('Correctly Predicted : ',correct_count)
print('Wrongly Predicted : ',total-correct_count)
```
